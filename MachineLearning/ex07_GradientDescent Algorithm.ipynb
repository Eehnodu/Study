## 1.1 학습데이터셋 생성

```python
import numpy as np
```

```python
x = np.random.uniform(low = 0.0, high = 1.0, size = 100)
x # 입력특성(공부시간)
```

```python
# 실제정답(성적)
y = 7 * x - 5
y
```

## 1.2 경사하강법 구현

```python
epoch = 500 # 가중치와 절편을 업데이트하는 횟수(반복횟수)
learning_rate = 1.0 # 학습률
w = np.random.uniform(low = 0.0, high = 1.0)
b = np.random.uniform(low  = 0.0, high = 1.0)
print("초기 가중치 :",  w)
print("초기 절편 :", b)
```

```python
for i in range(epoch):
    # 1. 가정에 대한 예측값 추출
    y_pre = w * x + b # 예측값
    # 2. 예측과 실제의 차이를 계산(loss function)
    MSE = ((y_pre - y) ** 2).mean()
    print("MSE : ",MSE)
    if MSE < 0.001 : # loss 값이 일정기준 이하일 때 중지 -> 자원가 시간을 아낌
        break
    # 3. 경사하강법을 이용한 방향(기울기) 계산구간 -> 편미분을 이용해 도출
    w = w - (learning_rate * ((y_pre - y)*x).mean())
    b = b - (learning_rate * ((y_pre - y)*x).mean())
    print("epoch : ", i, "w : ", w, "b : ",b)
```

### 입력특성이 2개 일때

```python
x1 = np.random.uniform(low = 0.0, high = 1.0, size = 100)
x2 = np.random.uniform(low = 0.0, high = 1.0, size = 100)
y = 10 * x1 - -5 * x2 + 5
```

```python
epoch = 500 # 가중치와 절편을 업데이트하는 횟수(반복횟수)
learning_rate = 100.0 # 학습률
w1 = np.random.uniform(low = 0.0, high = 1.0)
w2 = np.random.uniform(low = 0.0, high = 1.0)
b = np.random.uniform(low  = 0.0, high = 1.0)

print("초기 가중치 :",  w1)
print("초기 가중치 :",  w2)
print("초기 절편 :", b)
```

```python
for i in range(epoch):
    # 1. 가정에 대한 예측값 추출
    y_pre = w1 * x1 +  w2 * x2 + b # 예측값
    # 2. 예측과 실제의 차이를 계산(loss function)
    MSE = ((y_pre - y) ** 2).mean()
    print("MSE : ",MSE)
    if MSE < 0.001 : # loss 값이 일정기준 이하일 때 중지 -> 자원가 시간을 아낌
        break
    # 3. 경사하강법을 이용한 방향(기울기) 계산구간 -> 편미분을 이용해 도출
    w1 = w1 - (learning_rate * ((y_pre - y)*x1).mean())
    w2 = w2 - (learning_rate * ((y_pre - y)*x2).mean())
    b = b - (learning_rate * ((y_pre - y)*x).mean())
    print("epoch : ", i, "w1 : ", w1, "w2 : ", w2, "b : ",b)
```

## sklearn의 확률적 경사하강법 클래스 활용하기

```python
from sklearn.linear_model import SGDRegressor # 확률적 경사하강법 선형회귀
```

```python
sgd = SGDRegressor(eta0=0.01, # 학습률 
                   max_iter = 1000, # 최대반복횟수
                   epsilon = 0.1, # 학습을 진행하는 최소 오차 값
                   n_iter_no_change = 5, # 모델의 개선여부를 기다려주는 횟수
                   verbose=1
                  ) 
```

```python
sgd.fit(x.reshape(-1,1),y)
# loss -> MSE
```
