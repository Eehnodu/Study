```python
#  
import warnings
warnings.filterwarnings('ignore')
```

```python
from sklearn import datasets
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# opennml   
boston_data=datasets.fetch_openml(data_id=531)
boston_data
```

```python
 #  
#  
X=pd.DataFrame(boston_data.data, columns=boston_data.feature_names)
X.head()
```

```python
y=boston_data.target
y
```

```python
#     
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,
                                              random_state=423,
                                              test_size=0.2)
```

```python
X_train
```

#### LSTAT  LinearRegression 

```python
from sklearn.linear_model import LinearRegression
```

```python
plt.figure(figsize=(10,5))#

# Plot the training data
plt.scatter(X_train['LSTAT'], y_train)

# Fit the linear regression model
model = LinearRegression()
model.fit(X_train[['LSTAT']], y_train)
test_x=np.linspace(0,40,40).reshape(-1,1)
# Predict on the test data
pre = model.predict(test_x)

# Plot the regression line based on the test data predictions
plt.scatter(test_x,pre)

# Show the plot
plt.show()
```

```python
#    ->  
X_train['LSTAT x LSTAT']= X_train['LSTAT']*X_train['LSTAT']
```

```python
plt.figure(figsize=(10,5))#

# Plot the training data
plt.scatter(X_train['LSTAT'], y_train)

# Fit the linear regression model
model = LinearRegression()
model.fit(X_train[['LSTAT x LSTAT','LSTAT']], y_train)
test_x=np.linspace(0,40,40).reshape(-1,1)
test_x_2=test_x**2
text_x_concat=np.concatenate([test_x_2,test_x],axis=1)
# Predict on the test data
pre = model.predict(text_x_concat)

# Plot the regression line based on the test data predictions
plt.scatter(test_x,pre)

# Show the plot
plt.show()
```

```python
#      
columns=X_train.columns
columns
```

```python
X_train=X_train.astype(dtype='float64')
X_test=X_test.astype(dtype='float64')
```

```python
for c in columns:
    for c2 in columns:
        X_train[c+'x'+c2]=X_train[c]*X_train[c2]
```

```python
X_train.shape
```

```python
#   
X_test['LSTAT x LSTAT']=X_test['LSTAT']*X_test['LSTAT']
for c in columns:
    for c2 in columns:
        X_test[c+'x'+c2]=X_test[c]*X_test[c2]
```

```python
X_test.shape
```

```python
from sklearn.linear_model import Lasso # L1  
from sklearn.linear_model import Ridge # L2  
from sklearn.model_selection import cross_val_score #   
```

```python
linear_model=LinearRegression()
lasso_model=Lasso()
ridge_model=Ridge()
```

```python
#     3  
linear_score=cross_val_score(linear_model,X_train,y_train,cv=5)
lasso_score=cross_val_score(lasso_model,X_train,y_train,cv=5)
ridge_score=cross_val_score(ridge_model,X_train,y_train,cv=5)
```

```python
print('linear_score : ',linear_score.mean())
print('lasso_score : ',lasso_score.mean())
print('ridge_score : ',ridge_score.mean())
```

```python
#    
alpha_list = [0.0001, 0.001, 0.01, 0.1, 1, 10, 100, 1000, 10000]
lasso_score = []
ridge_score = []
for a in alpha_list:
    l = Lasso(alpha=a)
    r = Ridge(alpha=a)
    l_score = cross_val_score(l, X_train, y_train, cv = 5).mean()
    r_score = cross_val_score(r, X_train, y_train, cv = 5).mean()
    lasso_score.append(l_score)
    ridge_score.append(r_score)
```

```python
plt.figure(figsize=(10,5))
plt.plot(range(len(alpha_list)), lasso_score, label='Lasso score')
plt.plot(range(len(alpha_list)), ridge_score, label='Ridge score')
plt.legend()
plt.xticks(range(len(alpha_list)))
plt.grid()
plt.show()
```
