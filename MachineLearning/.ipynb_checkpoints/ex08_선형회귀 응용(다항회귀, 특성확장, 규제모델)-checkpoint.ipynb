- knn 모델은 이웃의 수를 줄이면 복잡도가 올라가고 이웃의 수를 늘리면 복잡도가 낮아진다.
<br>
- tree 모델은 depth의 깊이가 얇으면 복잡도는 낮아지고 깊이가 갚어지면 복잡도가 올라간다.
<br>
- LinearRegression모델은 따로 하이퍼 파라미터(ex depth,n_neughbor)가 없어 복잡도를 제어할 방법이 없어지기 때문에 과대적합 되기 쉽다.

### 목표
- 선형회귀 모델의 적합성을 높이기 위한 다항회귀 기법을 이해 할 수 있다
- 선형회귀 모델의 성능을 높이기 위한 특성확장 기법을 사용 할 수 있다.
- 과대적합을 방지하기 위한 규제모델을 이해하고 사용해보자.

#### 보스턴 데이터 로딩

```python
# 경고창 제거
import warnings
warnings.filterwarnings('ignore')
```

```python
from sklearn import datasets
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# opennml 사이트에서 보스턴데이터셋 다운로드
boston_data=datasets.fetch_openml(data_id=531)
boston_data
```

```python
 # 데이터셋 분리
# 문제와 답
X=pd.DataFrame(boston_data.data, columns=boston_data.feature_names)
X.head()
```

```python
y=boston_data.target
y
```

```python
# 훈련 데이터와 테스트 데이터로 분리
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,
                                              random_state=423,
                                              test_size=0.2)
```

```python
X_train
```

#### LSTAT와 주택가격을 LinearRegression으로 모델링

```python
from sklearn.linear_model import LinearRegression
```

```python
plt.figure(figsize=(10,5))#가로세로비율

# Plot the training data
plt.scatter(X_train['LSTAT'], y_train)

# Fit the linear regression model
model = LinearRegression()
model.fit(X_train[['LSTAT']], y_train)
test_x=np.linspace(0,40,40).reshape(-1,1)
# Predict on the test data
pre = model.predict(test_x)

# Plot the regression line based on the test data predictions
plt.scatter(test_x,pre)

# Show the plot
plt.show()
```

#### 다항회귀
- 데이터의 패턴에 따라서 직선형태의 선형모델이 아니라 곡선의 모델링이 필요할 수 있다
- 단 과대적합 여부는 별도로 판단해야함
- 다항회귀를 하는 방법은 특성을 확장하면 된다. -> 고차항을 1차에서 2차/3차 등으로 올려줘야 함

```python
# 입력특성의 차수를 증가 -> 특성 확장
X_train['LSTAT x LSTAT']= X_train['LSTAT']*X_train['LSTAT']
```

```python
plt.figure(figsize=(10,5))#가로세로비율

# Plot the training data
plt.scatter(X_train['LSTAT'], y_train)

# Fit the linear regression model
model = LinearRegression()
model.fit(X_train[['LSTAT x LSTAT','LSTAT']], y_train)
test_x=np.linspace(0,40,40).reshape(-1,1)
test_x_2=test_x**2
text_x_concat=np.concatenate([test_x_2,test_x],axis=1)
# Predict on the test data
pre = model.predict(text_x_concat)

# Plot the regression line based on the test data predictions
plt.scatter(test_x,pre)

# Show the plot
plt.show()
```

#### 규제모델 실습
- LinearRegression은 모델의 복잡도를 제어할 방법이 업다 -> 과대적합이 쉽다
- 과댈저합을 일반화하기 위해서 가중치(w) 값을 제어하는 모델 정규화 기법을 만듦
- L1(Lasso) : 모든 가중치에 동일한 힘으로 규제를 가하는 방식
- L2(Ridge) : 가중치의 크기에 비례하여 규제를 가하는 방식

```python
# 과대적합을 유도하기 위해서 특성확장을 과도하게 진행
columns=X_train.columns
columns
```

```python
X_train=X_train.astype(dtype='float64')
X_test=X_test.astype(dtype='float64')
```

```python
for c in columns:
    for c2 in columns:
        X_train[c+'x'+c2]=X_train[c]*X_train[c2]
```

```python
X_train.shape
```

```python
# 테스트 데이터 특성확장
X_test['LSTAT x LSTAT']=X_test['LSTAT']*X_test['LSTAT']
for c in columns:
    for c2 in columns:
        X_test[c+'x'+c2]=X_test[c]*X_test[c2]
```

```python
X_test.shape
```

```python
from sklearn.linear_model import Lasso # L1 규제가 적용
from sklearn.linear_model import Ridge # L2 규제가 적용
from sklearn.model_selection import cross_val_score # 교차 검증 함수
```

```python
linear_model=LinearRegression()
lasso_model=Lasso()
ridge_model=Ridge()
```

```python
# 특성확장시킨 훈련 데이터를 이용해 3개 모델 교차검즘
linear_score=cross_val_score(linear_model,X_train,y_train,cv=5)
lasso_score=cross_val_score(lasso_model,X_train,y_train,cv=5)
ridge_score=cross_val_score(ridge_model,X_train,y_train,cv=5)
```

```python
print('linear_score : ',linear_score.mean())
print('lasso_score : ',lasso_score.mean())
print('ridge_score : ',ridge_score.mean())
```

```python
# 규제 모델 튜닝 실습
alpha_list = [0.0001, 0.001, 0.01, 0.1, 1, 10, 100, 1000, 10000]
lasso_score = []
ridge_score = []
for a in alpha_list:
    l = Lasso(alpha=a)
    r = Ridge(alpha=a)
    l_score = cross_val_score(l, X_train, y_train, cv = 5).mean()
    r_score = cross_val_score(r, X_train, y_train, cv = 5).mean()
    lasso_score.append(l_score)
    ridge_score.append(r_score)
```

```python
plt.figure(figsize=(10,5))
plt.plot(range(len(alpha_list)), lasso_score, label='Lasso score')
plt.plot(range(len(alpha_list)), ridge_score, label='Ridge score')
plt.legend()
plt.xticks(range(len(alpha_list)))
plt.grid()
plt.show()
```
