# 문제4

```python
#10
score_list = [["박매일", 89, 78, 99],
             ["서대희", 67, 88, 77],
             ["강예진", 100, 55, 96],
              ["차현석", 75, 100, 60]
             ]
python, java, db, avg = 0 ,0 ,0, 0
grade = ""
print("============================================")
print("이름\tPython\tJava\tDB\t평균\t학점")
print("============================================")
for i in score_list:
    avg = (i[1]+i[2]+i[3])//3
    if(avg >= 90):
        grade = "A"
    elif (avg >= 80):
        grade = "B"
    elif (avg >= 70):
        grade = "C"
    elif (avg >= 60):
        grade = "D"
    else:
        grade = "F"
    print(f'{i[0]}\t{i[1]}\t{i[2]}\t{i[3]}\t{avg}\t{grade}')
    python += i[1]
    java += i[2]
    db += i[3]
    avg = 0
print("============================================")
python = python // len(score_list)
java = java // len(score_list)
db = db // len(score_list)
print(f'평균\t{python}\t{java}\t{db}')
```

# 문제5

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

X = [2,3,4,5,6,7,8]
Y1 = [5,7,3,9,1,8,6]
Y2 = [9,5,7,2,3,7,1]

plt.figure(figsize=(9,5))

plt.plot(X, Y1, label = 'Y1', c='red')
plt.plot(X, Y2, label = 'Y2')

plt.legend()
plt.xlabel('X')
plt.ylabel('Y')
plt.grid()
plt.show()
```

# 문제6

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

accident = pd.read_csv('../data/Traffic_Accident_2017.csv', encoding = 'euc-kr')
accident.head(3)
```

```python
ctc = accident[accident['사고유형_대분류'] == '차대사람']
ctc_night = ctc[ctc['주야'] == '야간']
ctc_night
```

```python
ctc_night['사고유형_대분류'].value_counts()
```

```python
tmp = ctc_night['요일'].value_counts()
y = tmp[['월','화','수','목','금','토','일']]
y
```

```python
plt.rcParams['font.family'] = 'Malgun Gothic'
x = y.index
plt.figure(figsize=(9,5))
plt.bar(x,y)
plt.title('요일별 야간 차대사람 교통사고 건수')
plt.xlabel('요일')
plt.ylabel('사고 건수')
plt.ylim(0,200)
plt.show()
```
