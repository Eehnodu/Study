```python
import pandas as pd
import numpy as np
exam = pd.read_csv('../data/exam.csv')
exam
```

```python
exam.info()
```

```python
exam.describe()
```

```python
exam['total'] = exam['math'] + exam['english'] + exam['science']
exam
```

```python
exam['avg'] = round((exam['math'] + exam['english'] + exam['science'])/3,2)
exam
```

```python
exam['total_loc'] = exam.loc[:,'math':'science'].sum(axis = 1)
exam['avg_loc'] = round(exam.loc[:,'math':'science'].mean(axis = 1),2)
exam
```

```python
exam['total_iloc'] = exam.iloc[:,2:5].sum(axis = 1)
exam['avg_iloc'] = round(exam.iloc[:,2:5].mean(axis = 1),2)
exam
```

```python
exam['result'] = np.where(exam['avg'] >= 60, 'pass', 'fail')
exam
```

```python
def grade(row):
    mean = row['avg']
    if mean >= 80:
        return 'A'
    elif mean >= 70:
        return 'B'
    elif mean >= 67:
        return 'C'
    else :
        return 'F'
```

```python
exam['grade'] = exam.apply(grade, axis=1)
exam
```

```python
exam.groupby(['nclass','result']).agg(result_cnt = ('result','count')).sort_values(by = 'result', ascending=False)
```

```python
exam.groupby(['nclass','grade']).agg(grade_cnt = ('grade','count'))
```

```python
exam.groupby(['nclass','result']).agg(result_cnt = ('result','count')).sort_index(ascending=False).sort_values(by = 'nclass', ascending=True)
```
