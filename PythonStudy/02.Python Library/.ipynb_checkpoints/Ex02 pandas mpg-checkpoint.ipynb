```python
import numpy as np
mpg_copy['test결과'] = np.where(mpg_copy['연비평균'] >= 20, "pass", "fail")
mpg_copy
```

```python
mpg_copy['test결과2'] = np.where(mpg_copy['연비평균'] >= 25, "best", np.where(mpg_copy['연비평균'] >= 20, "pass", "fail"))
mpg_copy
```

```python
def grade(row):
    mean = row['연비평균']
    if mean >= 25:
        return 'A'
    elif mean >= 20:
        return 'B'
    elif mean >= 17:
        return 'C'
    elif mean >= 15:
        return 'D'
    else :
        return 'F'
```

```python
mpg_copy.apply(grade, axis=1)
```

```python
mpg_copy['Grade'] = mpg_copy.apply(grade, axis=1)
mpg_copy
```

```python
mpg_copy.groupby(['제조사','Grade'], as_index=False).agg(등급합계 = ('Grade', 'count'))
```
