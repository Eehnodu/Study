- 브라우저 대신 파이썬에서 페이지 정보를 요청 할 때 사용

```python
# 1. 네이버에게 메인 페이지를 요청
# req를 통해서 네이버 페이지 가져와
# resopnse [200, 300] -> 통신에 성공했습니다.
# resopnse [400] -> 통신에 실패 (클라이언트의 요청의 문제)
# resopnse [500] -> 통신에 실패 (서버에 문제가 있다)
import requests as req
from bs4 import BeautifulSoup
res = req.get("https://www.naver.com")
soup = BeautifulSoup(res.content, 'html.parser')
title = soup.find('title')
print(title.get_text())
```

# Melon사이트의 정보를 요청

```python
res1 = req.get("https://www.melon.com/chart/index.htm")
print(res1)

# response [200,300] > 통신에 성공했습니다.
# response[400] > 통신에 실패(클라이언트의 요청의 문제)
# response[500] > 통신에 실패(서버의 문제.)
```

```python
import requests as req
from bs4 import BeautifulSoup as bs

url = "https://www.melon.com/chart/index.htm"
header = {"User-Agent":"Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/100.0.4896.75 Safari/537.36"}
resp = req.get(url, headers=header)
print(resp)

# 응답코드가 406 > 요청하는 클라이언트 문제
# 멜론서버가 요청하는 주체가 브라우저가 아님을 인지
# 브라우저인 척 속여서 요청을 보내야 한다.
# 내가 원하는 사이트가 응답이 406이 나오면 이 방법을 무조건 시도해봐라!
```

```python
soup = bs(resp.text, "lxml")
title_list = []
# soup.find_all('a', {"class" : "btn button_icons type03 song_info"})
for title in soup.find_all('a', {"class" : "btn button_icons type03 song_info"}):
    title_list.append(title.get('title')[:-3])
print(title_list)
```

# beautifulSoup 라이브러리
- req로 받아온 텍스트 데이터를 컴퓨터가 이해 할 수 있게 HTML로 변환 시켜주는 라이브러리

```python
from bs4 import BeautifulSoup as bs
```

```python
soup = bs(res.text,"lxml")
```

```python
for title in soup.find_all('a', {'class' : 'news_tit'}):
    print(title.get('title'))
```

```python
for title in soup.select("a.news_tit"):
    print(title.get('title'))
    
# + .select_one >한개만 리턴
```
