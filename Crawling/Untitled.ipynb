```python
from wordcloud import WordCloud
from collections import Counter

from konlpy.tag import Okt
from konlpy.tag import Komoran

from PIL import Image
import matplotlib.pyplot as plt
import numpy as np
```

```python
with open('C:\\Users\\smhrd\\Desktop\\Linux\\Virtual-Hadoop\\04.WordCount\\대한민국헌법.txt', 'r', encoding='utf-8') as f:
    text = f.read()
```

```python
text
```

```python
okt = Okt()
```

```python
# 명사만 추출
nouns = okt.nouns(text) 

words = [n for n in nouns if len(n) > 1] # 단어의 길이가 1개인 것은 제외
```

```python
# 위에서 얻은 words를 처리하여 단어별 빈도수 형태의 딕셔너리 데이터를 구함
wcnt = Counter(words) 
wc = WordCloud(font_path='malgun', width=400, height=400, scale=2.0, max_font_size=250)
gen = wc.generate_from_frequencies(wcnt)
plt.figure()
plt.imshow(gen)
```

```python
wc.to_file('법전_워드클라우드.png')
```

```python
with open('C:\\Users\\smhrd\\Desktop\\Crawling\\역대대통령신년사.csv', 'r', encoding='utf-8') as f:
    text = f.read()
```

```python
text
```

```python
# 데이터를 딕셔너리로 변환
word_freq = {}
for line in text.split('\n'):
    word, freq = line.split('\t')
    word_freq[word] = int(freq)

word_freq
```

```python
# 워드 클라우드 생성
wc = WordCloud(font_path='malgun', width=400, height=400, scale=2.0, max_font_size=250)
gen = wc.generate_from_frequencies(word_freq)

# 워드 클라우드 그리기
plt.imshow(gen, interpolation="bilinear")
plt.axis("off")  # 축 제거
plt.show()
```

```python
import os

folder_path = "C:\\Users\\smhrd\\Downloads\\대통령신년사\\"

# 폴더 내의 모든 파일 목록 가져오기
file_list = os.listdir(folder_path)

# 폴더 내의 모든 텍스트 파일 읽기
for file_name in file_list:
    if file_name.endswith('.txt'):  # txt 파일인지 확인
        file_path = os.path.join(folder_path, file_name)
        with open(file_path, 'r', encoding='utf-8') as f:
            text = f.read()
```

```python
okt = Okt()
```
